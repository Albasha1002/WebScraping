```python
import requests
```

```python
from bs4 import BeautifulSoup as bs
```

```python
r=requests.get('https://keithgalli.github.io/web-scraping/webpage.html')
```

```python
soup= bs(r.content)
```

```python
soup.find("span")
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
results=soup.find_all(["h1", "h2"])
```

```python
results[0].text
```

```python
records=[]
for result in results:
   records.append(result.text[0:])
```

```python
records
```

```python
import pandas as pd
```

```python
Heading=pd.DataFrame(records, columns=['H1 command'])
```

```python
Heading.to_csv('Heading', index=False)
```

```python
paragrap=soup.find_all('p')
```

```python
para=[]
for result in paragrap:
  para.append(result.text)
```

```python
para
```

```python
paragraph=pd.DataFrame(para, columns=['Paragraph'])
```

```python
paragraph.to_csv("Paragraph", encoding='utf-8')
```
